# Data structure: [ restaurant_id, 2021, 2022, 2023, 2024]

```python
import numpy as np
```

```python
sales_data = np.array([
    [1,150000,180000,220000,250000],
    [2,120000,140000,160000,190000],
    [3,200000,210000,240000,270000],
    [4,180000,210000,240000,270000],
    [5,160000,185000,205000,230000]
])

print("--- Zomato sales analysis ---")
print("\n Sales data shape :", sales_data.shape)
print("First 3 restrorent data :\n", sales_data[:3])
```

# Total sales per year

```python
yearly_total = np.sum(sales_data[:,1:],axis=0)
print(yearly_total)

min_sales = np.min(sales_data[:,1:], axis=1)
print("Min sales",min_sales)
```

```python
max_sales_per_year = np.max(sales_data[:,1:], axis=0)
print("Max sales per year :",max_sales_per_year)
```
