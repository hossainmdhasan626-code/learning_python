```python
import numpy as np
import time
```

```python
arr_1d = np.array([1,2,3,4,5])
print("1D array: ",arr_1d)

arr_2d = np.array([[1,2,3],[4,5,6]])
print("2D array: ",arr_2d)
```

### List vs numpy array

```python
py_list = [1,2,3]
print("Python list multiplication ", py_list * 2)

np_array = np.array([1,2,3])
print("Python array multiplication ", np_array * 2)
```

```python
start = time.time()
py_list = [i*2 for i in range(1000)]
print("\n List operation time: ",time.time() - start)


start = time.time()
np_array = np.arange(1000) * 2
print("\n Array operation time: ",time.time() - start)
```

### creating array from scratch

```python
zeros = np.zeros((3,4))
print("zeros array: \n",zeros)

ones = np.ones((2,3))
print("ones array: \n",ones)

full = np.full((2,2),7)
print("full array: \n",full)

random = np.random.random((2,3))
print("random array: \n",random)

sequence = np.arange(0,10,2)
print("sequence array: \n",sequence)
```

### Vector, Matrix and Tensor

### Arrat properties

### Array Reshaping

```python
arr = np.arange(12)
print("Original array ",arr)

reshaped = arr.reshape((3,4))
print("Reshaped array ",reshaped)

flattened = reshaped.flatten()
print("Flattened array ",flattened)

# ravel ( original e change ane )
raveled = reshaped.ravel()
print("Raveled array ",raveled)

transpose = reshaped.T
print("Transposed array ",transpose)
```

```python

# একটি ৩x৩ ম্যাট্রিক্স বা টেবিল তৈরি করলাম
# ইন্ডেক্স মনে রাখার ট্রিক:
# সারি (Rows)    ->   0 নম্বর: [10, 20, 30] | 1 নম্বর: [40, 50, 60] | 2 নম্বর: [70, 80, 90]
# কলাম (Columns) ->   0 নম্বর: [10,40,70]   | 1 নম্বর: [20,50,80]   | 2 নম্বর: [30,60,90]

arr = np.array([
    [10, 20, 30],  # সারি ০
    [40, 50, 60],  # সারি ১
    [70, 80, 90]   # সারি ২
])

print("--- ১. নির্দিষ্ট একটি উপাদান (Single Indexing) ---")
# নিয়ম: arr[row, column]
# এখানে row = 1 (মানে ২য় সারি), column = 2 (মানে ৩য় কলাম)
element = arr[1, 2] 
print("নির্দিষ্ট উপাদান:", element) # আউটপুট আসবে: 60
print("\n")


print("--- ২. ডাটা কেটে আনা (Slicing) ---")
# আমাদের টার্গেট কোড: arr[0:2, 1:3]
# স্লাইসিং এর গোল্ডেন রুল: [start : end] মানে হলো start থেকে (end - 1) পর্যন্ত।

# ১ অংশ [0:2] -> এটি সারির (Row) জন্য।
# হিসেব: শুরু হবে ০ থেকে, শেষ হবে (২ - ১) = ১ এ।
# অর্থাৎ, সে শুধু 'সারি ০' এবং 'সারি ১' কে কেটে নেবে। (সারি ২ বাদ!)

# ২য় অংশ [1:3] -> এটি কলামের (Column) জন্য।
# হিসেব: শুরু হবে ১ থেকে, শেষ হবে (৩ - ১) = ২ এ।
# অর্থাৎ, সে শুধু 'কলাম ১' এবং 'কলাম ২' কে কেটে নেবে। (কলাম ০ বাদ!)

sliced_data = arr[0:2, 1:3]

print("কেটে আনা নতুন ম্যাট্রিক্স:")
print(sliced_data)
# আউটপুট আসবে:
# [[20 30]     <- সারি ০ এর কলাম ১ ও কলাম ২
#  [50 60]]    <- সারি ১ এর কলাম ১ ও কলাম ২
```

```python
# মনে করুন এটি একটি ই-কমার্স অ্যাপের কিছু প্রোডাক্টের দাম (Prices)
prices = np.array([50, 120, 80, 200, 45, 300])

print("--- ১. ফিল্টারিং করার ম্যাজিক ---")

# আমাদের শর্ত: আমরা শুধু ১০০ টাকার বেশি দামি প্রোডাক্টগুলো দেখতে চাই
# ব্যাকঅ্যান্ডে NumPy প্রতিটা উপাদানের জন্য True অথবা False এর একটা 'মাস্ক' বানায়
expensive_mask = prices > 100 
print("শর্তের মাস্কটি দেখতে এমন হয়:\n", expensive_mask)
# আউটপুট: [False  True False  True False  True]
# (কারণ ১২০, ২০০, এবং ৩০০ হলো ১০০-এর চেয়ে বড়, তাই ওগুলো True)

# এবার এই মাস্কটি অরিজিনাল অ্যারের ভেতর পাস করলেই, সে শুধু True ডাটাগুলো কেটে আনবে!
expensive_prices = prices[prices > 100]

print("\n১০০ টাকার বেশি দামি প্রোডাক্টের লিস্ট:")
print(expensive_prices) 
# আউটপুট আসবে: [120 200 300]
```

```python
# একটি সিম্পল ২-ডাইমেনশনাল ম্যাট্রিক্স নিলাম
matrix_data = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("--- ২. ডাটার হিসাব-নিকাশ ---")

# ১. পুরো ম্যাট্রিক্সের সবগুলো সংখ্যার যোগফল (Sum)
total_sum = np.sum(matrix_data)
print("সব সংখ্যার মোট যোগফল:", total_sum) # ১+২+৩+৪+৫+৬ = ২১

# ২. পুরো ম্যাট্রিক্সের গড় মান (Mean/Average)
average = np.mean(matrix_data)
print("ডাটার গড় মান (Average):", average) # ২১ কে ৬ দিয়ে ভাগ করলে ৩.৫ হয়

# ৩. পুরো ম্যাট্রিক্সের মধ্যে সবচেয়ে বড় সংখ্যা (Maximum)
max_value = np.max(matrix_data)
print("সবচেয়ে বড় সংখ্যা কোনটি?:", max_value) # আউটপুট আসবে: ৬

# ৪. পুরো ম্যাট্রিক্সের মধ্যে সবচেয়ে ছোট সংখ্যা (Minimum)
min_value = np.min(matrix_data)
print("সবচেয়ে ছোট সংখ্যা কোনটি?:", min_value) # আউটপুট আসবে: ১
```
